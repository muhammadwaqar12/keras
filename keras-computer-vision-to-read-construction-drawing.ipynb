```python
# Set the path to the dataset
# Make sure to update the data_dir variable to the path of your desired dataset
data_dir = '/kaggle/input/dataset/to/csv/file.csv'

# Define the size of the input images
image_height = 256
image_width = 256
channels = 3

# Create a data generator for the training set
train_data_generator = ImageDataGenerator(rescale=1./255,
                                           shear_range=0.2,
                                           zoom_range=0.2,
                                           horizontal_flip=True)
train_generator = train_data_generator.flow_from_directory(directory=os.path.join(data_dir, 'train'),
                                                           target_size=(image_height, image_width),
                                                           batch_size=32,
                                                           class_mode='categorical')

# Create a data generator for the validation set
val_data_generator = ImageDataGenerator(rescale=1./255)
val_generator = val_data_generator.flow_from_directory(directory=os.path.join(data_dir, 'val'),
                                                       target_size=(image_height, image_width),
                                                       batch_size=32,
                                                       class_mode='categorical')

# Create a data generator for the test set
test_data_generator = ImageDataGenerator(rescale=1./255)
test_generator = test_data_generator.flow_from_directory(directory=os.path.join(data_dir, 'test'),
                                                         target_size=(image_height, image_width),
                                                         batch_size=32,
                                                         class_mode='categorical')
```

```python
# Initialize the model
model = Sequential()

# Add the convolutional layer
model.add(Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(image_height, image_width, channels)))

# Add the pooling layer
model.add(MaxPooling2D(pool_size=(2, 2)))

# Add the flattening layer
model.add(Flatten())

# Add the fully connected layers
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=3, activation='softmax'))

# Print the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(train_generator,
                    epochs=10,
                    validation_data=val_generator)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(test_generator)
print('Test accuracy:', accuracy)
```
